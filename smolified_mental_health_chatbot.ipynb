## Local Inference on GPU
Model page: https://huggingface.co/smolify/smolified-mental-health-chatbot

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/smolify/smolified-mental-health-chatbot)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="smolify/smolified-mental-health-chatbot")
messages = [
    {"role": "user", "content": "I feel lonely and tired these days. I don't have the motivation to take care of myself either, like simple workout or eating good and right."},
]
pipe(messages)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("smolify/smolified-mental-health-chatbot")
model = AutoModelForCausalLM.from_pretrained("smolify/smolified-mental-health-chatbot")
messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
